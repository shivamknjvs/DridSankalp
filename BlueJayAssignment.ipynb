```python
import pandas as pd

# Read the CSV file
file_path = "/content/Assignment_Timecard.xlsx"
df = pd.read_excel(file_path)

# Print the column names
print("Column names:", df.columns)

df['Time'] = pd.to_datetime(df['Time'])
df['Time Out'] = pd.to_datetime(df['Time Out'])



# Function to check consecutive days worked
def consecutive_days(df):
    consecutive_days_threshold = 7
    print("\nEmployees who have worked for {} consecutive days:".format(consecutive_days_threshold))

    for name, group in df.groupby('Employee Name'):
        consecutive_days_count = group['Time'].nunique()

        if consecutive_days_count >= consecutive_days_threshold:
            print("Employee Name: {}".format(name))
            print("Position ID: {}".format(group['Position ID'].iloc[0]))
            print("Consecutive Days Worked: {}".format(consecutive_days_count))
            print("\n")

# Function to check time between shifts
def time_between_shifts(df):
    min_hours_between_shifts = 1
    max_hours_between_shifts = 10
    print("\nEmployees with less than {} hours but greater than {} hours between shifts:".format(max_hours_between_shifts, min_hours_between_shifts))

    for name, group in df.groupby('Employee Name'):
        time_between_shifts = group['Time'].diff().dt.total_seconds() / 3600  # convert seconds to hours
        violated_shifts = group[(time_between_shifts > min_hours_between_shifts) & (time_between_shifts < max_hours_between_shifts)]

        if not violated_shifts.empty:
            print("Employee Name: {}".format(name))
            print("Position ID: {}".format(group['Position ID'].iloc[0]))
            print("\n")

# Function to check single shift duration
def long_shifts(df):
    max_hours_per_shift = 14
    print("\nEmployees who have worked for more than {} hours in a single shift:".format(max_hours_per_shift))

    for name, group in df.groupby('Employee Name'):
        shift_duration = (group['Time Out'] - group['Time']).dt.total_seconds() / 3600  # convert seconds to hours
        long_shifts = group[shift_duration > max_hours_per_shift]

        if not long_shifts.empty:
            print("Employee Name: {}".format(name))
            print("Position ID: {}".format(group['Position ID'].iloc[0]))
            print("\n")

# Call the functions to print the results
consecutive_days(df)
time_between_shifts(df)
long_shifts(df)
```
